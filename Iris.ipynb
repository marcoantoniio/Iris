# Fazendo o processo inicial como importar as bibliotecas, conferir o dataset etc

```python
# Importando as bibliotecas que irei usar

import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# Importando o dataset Iris

iris = load_iris()
```

# Criando um DataFrame com os dados do Dataset para ficar melhor a visualização e entendimento dos dados

```python
# Organizando os dados dentro do dataframe

iris_df = pd.DataFrame(data=iris.data, columns = iris.feature_names)
```

```python
# Colocando duas colunas no dataset, uma com o nome das espécies e a outra com o seu respectivo número

iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
iris_df['label'] = iris.target
```

```python
# Verificando quantas linhas e colunas o dataset possui

iris_df.shape
```

```python
# Confirmando todos as especies do dataset e sua quantidade

fig, ax = plt.subplots(figsize=(16,9))

ax.bar(iris_df.species.unique(), iris_df.species.value_counts())
```

```python
# Plotando vários gráficos para entender melhor a distribuição de cada espécie

sns.pairplot(iris_df, hue='species')
```

```python
# Verificando quantos exemeplos eu tenho de cada espécie

iris_df['species'].value_counts()
```

```python
# Criando uma planilha excel com os dados do dataset

iris_df.to_excel('iris_df.xlsx')
```

# Criando o modelo que irá fazer a predição

```python
# Importando a biblioteca de machine learning

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
```

```python
# Separando o dataframe em dois, o X são os dados que o modelo irá usar para aprender sobre as espécies
# E o Y são os valores que eu desejo prever com o modelo


X = iris_df.drop(['species', 'label'], axis=1)
y = iris_df['label']
```

```python
# Fazendo a separação dos dados de treion e de teste, neste caso estou usando 60% para treinar o modelo
# E os outros 40% para testar o mesmo 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
# Criando o modelo e passando os parâmetros como o número de neuronios que terá nas camadas escocindas da rede neural
# Seu metodo de ativação, a solução para otimizar os pesos de cada neuronio e o máximo de iterações que o modelo terá com os dados

clf = MLPClassifier(hidden_layer_sizes=(10,5),
                   activation='relu',
                   solver='sgd',
                   learning_rate='constant',
                   tol=0.0000001,
                   max_iter=1000,
                   verbose=2,
                   alpha=1e-5,
                   random_state=42)
```

```python
# Preenchendo o modelo com os dados de treino

clf.fit(X_train, y_train)
```

```python
# Fazendo a predição com os dados de teste

pred_clf = clf.predict(X_test)
```

```python
# Importando a biblioteca para medir a acurácia do modelo

from sklearn.metrics import accuracy_score
```

```python
# Calculando a acurácia do modelo, esse valor pode variar entre 1 e 0, sendo 1 igual a 100%

accuracy_score(y_test, pred_clf)
```

```python
# Importando a biblioteca para salvar o modelo

from joblib import load, dump
```

```python
# Savlvando o modelo de forma que ele possa ser usado novamente no futuro sem precisar ser treinado novamente

dump(clf, 'clfiris.joblib')
```

```python
# Importando o modelo para testalo

modelo = load('clfiris.joblib')
```

```python
# Verificando as informações do modelo

modelo
```

```python
# Importando a planilha para testar o modelo

testes = pd.read_excel('iris_df.xlsx')
```

```python
# Deletando as colunas que possuem as respostas

testes = testes.drop(['species', 'label', 'Unnamed: 0'], axis=1)
```
